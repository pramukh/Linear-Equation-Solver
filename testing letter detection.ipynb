```python
import cv2 as cv
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
def disp_img(img):
    plt.xticks([])
    plt.yticks([])
    plt.imshow(img)
    plt.show()
```

```python
mpl.rcParams['figure.dpi']= 150
```

```python
img = cv.imread('random_tests.jpg')
```

```python
img.shape
```

```python
disp_img(img)
```

```python
gray = cv.cvtColor(img.copy(), cv.COLOR_BGR2GRAY)
blurred = cv.GaussianBlur(gray, (5, 5), 10)
thresh = cv.adaptiveThreshold(blurred, 255, 0, 1, 115, 1)

disp_img(thresh)
```

```python
kernel = np.ones((3,3),np.uint8)
dilation = cv.dilate(thresh,kernel,iterations = 1)
```

```python
disp_img(dilation)
```

```python
contours, hierarchy = cv.findContours(dilation, cv.RETR_EXTERNAL, cv.CHAIN_APPROX_SIMPLE)
contours = sorted(contours, key=lambda ctr: cv.boundingRect(ctr)[0])
rects = [cv.boundingRect(ctr) for ctr in contours]
```

```python
contours
```

```python
rects
```

```python
for ctr in contours:
    x1 = ctr[0]
    x2 = ctr[0] + ctr[2]
    
```

```python
ouput_img = img.copy()
extracted_digits = []
pad = 1
for (x, y, w, h) in rects:
    extracted_digits.append(dilation.copy()[y-pad: y + h+pad, x-pad: x + w+pad])
    rect = cv.rectangle(ouput_img, (x, y), (x + w, y + h), (0, 0, 255), 3)
```

```python
disp_img(ouput_img)
```

```python
ouput_img2 = img.copy()
for cnt in contours:
    (x,y),radius = cv.minEnclosingCircle(cnt)
    center = (int(x),int(y))
    radius = int(radius)
    cv.circle(ouput_img2,center,radius,(0,255,0),2)
```

```python
disp_img(ouput_img2)
```

```python
output_img3 = img.copy()
for cnt in contours:
    rows,cols = output_img3.shape[:2]
    [vx,vy,x,y] = cv.fitLine(cnt, cv.DIST_L2,0,0.01,0.01)
    lefty = int((-x*vy/vx) + y)
    righty = int(((cols-x)*vy/vx)+y)
    cv.line(output_img3,(cols-1,righty),(0,lefty),(0,255,0),2)
```

```python
disp_img(output_img3)
```

```python
output_img4 = img.copy()
for cnt in contours:
    ellipse = cv.fitEllipse(cnt)
    cv.ellipse(output_img4,ellipse,(0,255,0),2)
```

```python
disp_img(output_img4)
```
